```python
age=[1,3,28,27,25,92,30,39,40,50,26,24,29,94]
```

```python
df=pd.DataFrame(age)
```

```python
df
```

```python
sns.boxplot(df)
```

```python
sns.scatterplot(data=df)
```

```python
q1=df.quantile(0.25)
q2=df.quantile(0.5)
q3=df.quantile(0.75)
```

```python
iqr=q3-q1
iqr
```

```python
Q1=np.percentile(df,25)
Q2=np.percentile(df,50)
Q3=np.percentile(df,75)
IQR=Q3-Q1
IQR
```

```python
l=Q1-1.5*IQR
h=Q3+1.5*IQR
```

```python
l
```

```python
h
```

```python
outliers=[x for x in age if x<l or x>h]
print("Q1:",Q1)
print("Q2:",Q2)
print("Q3:",Q3)
print("IQR:",IQR)
print("Lower whisker:",l)
print("Upper whisker:",h)
print("Outliers:",outliers)
```

```python
df=df[((df>=l)&(df<=h))]
df
```

```python
df=df.dropna()
df
```

```python
sns.boxplot(data=df)
```

```python
sns.scatterplot(data=df)
```

```python
data=[1,2,2,2,3,1,1,15,2,2,2,3,1,1,2]
mean=np.mean(data)
std=np.std(data)
print("mean of the dataset is ",mean)
print("standard deviation of the dataset is ",std)
```

```python
threshold=3
outlier=[]
for i in data:
  z=(i-mean)/std
  if z>threshold:
    outlier.append(i)
print("outlier in the dataset is ",outlier)
```
